```python
import opendatasets as od

od.download("https://www.kaggle.com/datasets/chaitu20/ipl-dataset2008-2025")
```

```python
import pandas as pd
df=pd.read_csv("/content/ipl-dataset2008-2025/IPL.csv")
```

```python
print(df.head())
```

```python
df1=df[['innings','batting_team','bowling_team','over','ball','team_runs','team_balls','team_wicket','venue','city','superover_winner','match_won_by','runs_target','ball_no']]
```

```python
df1.head()
```

```python
df2=df1.query("`innings` == 2")
df2.head()
```

```python
df2=df2.assign(runs_left=df2['runs_target']-df2['team_runs'])
df2.head()
#runs left
```

```python
df2=df2.assign(balls_left=120-df2['team_balls'])
df2.head()
#balls left
```

```python
#wickets left
df2=df2.assign(wickets_left=10-df2['team_wicket'])
df2.head()
```

```python
#crr
df2 = df2.assign(crr = df2['team_runs']*6/df2['team_balls'])
df2.head()
```

```python
#rrr
df2=df2.assign(rrr=df2['runs_left']*6/df2['balls_left'])
df2.head()
```

```python
df3=df2[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','runs_target','crr','rrr','match_won_by','superover_winner']]
df3.head()
```

```python
df3['match_won_by'] = df3.apply(lambda row: 1 if row['match_won_by'] == row['batting_team'] else 0, axis=1)
df3.head()
```

```python
df3=df3.rename(columns={'match_won_by':'result'})
df3.head()
```

```python
df3['superover_winner'].fillna(0, inplace=True)
df3=df3.query("`superover_winner` == 0")

df3.head()
```

```python
df3 = df2[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left',
           'wickets_left', 'runs_target', 'crr', 'rrr', 'match_won_by', 'superover_winner']].copy()

teams_to_remove = ['Pune Warriors', 'Rising Pune Supergiant', 'Kochi Tuskers Kerala']

df3 = df3[~df3['batting_team'].isin(teams_to_remove)]
df3 = df3[~df3['bowling_team'].isin(teams_to_remove)]

df3.head()
```

```python
df3['batting_team'] = df3['batting_team'].replace('Deccan Chargers','Sunrisers Hyderabad')
df3['batting_team'] = df3['batting_team'].replace('Delhi Daredevils','Delhi Capitals')
df3['batting_team'] = df3['batting_team'].replace('Gujarat Lions','Gujarat Titans')
df3['batting_team'] = df3['batting_team'].replace('Royal Challengers Bengaluru','Royal Challengers Bangalore')
df3['batting_team'] = df3['batting_team'].replace('Kings XI Punjab','Punjab Kings')

df3['bowling_team'] = df3['bowling_team'].replace('Deccan Chargers','Sunrisers Hyderabad')
df3['bowling_team'] = df3['bowling_team'].replace('Delhi Daredevils','Delhi Capitals')
df3['bowling_team'] = df3['bowling_team'].replace('Gujarat Lions','Gujarat Titans')
df3['bowling_team'] = df3['bowling_team'].replace('Royal Challengers Bengaluru','Royal Challengers Bangalore')
df3['bowling_team'] = df3['bowling_team'].replace('Kings XI Punjab','Punjab Kings')

df3['city'] = df3['city'].replace('Bengaluru','Bangalore')

df3.head()
```

```python
final_df = df3[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left', 'wickets_left', 'runs_target', 'crr', 'rrr', 'match_won_by']].copy()

final_df['match_won_by'] = final_df.apply(lambda row: 1 if row['match_won_by'] == row['batting_team'] else 0, axis=1)

final_df = final_df.rename(columns = {'match_won_by':'result', 'wickets_left': 'wickets', 'runs_target': 'total_runs_x'})
```

```python
final_df = final_df.rename(columns = {'wickets_left' : 'wickets', 'runs_target' : 'total_runs_x'})
```

```python
final_df.sample(20)
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
final_df = final_df[final_df['balls_left'] != 120]
```

```python
X = final_df.iloc[:, :-1]
y = final_df.iloc[:, -1]

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output = False, drop = 'first'), ['batting_team', 'bowling_team', 'city'])
]
, remainder = 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe = Pipeline(steps = [
    ('step1', trf),
    ('step2', LogisticRegression(solver = 'liblinear'))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {f1}")
```

```python
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred)
print(f"Precision Score: {precision}")
```
